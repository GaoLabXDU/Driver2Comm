# Breast cancer data example
## import package

```python
import Driver2Comm as dc
import pandas as pd
import os
```

## read driver gene information of each patient

```python
patient_driver_info = pd.read_csv('./data/Driver2Comm_input/pseudo_patient_driver.csv')
patient_driver_info
```

## read formulated CCC data of each patient

```python
c2c_network = dc.read_c2c_network(os.path.join('./data/Driver2Comm_input/c2c_network.data'))
```

## construct Driver2Comm object and run it

```python
model = dc.Driver2Comm(
    c2c_network = c2c_network,
    patient_driver_info=patient_driver_info,
    minsup= 14,
    association_test_threshold=0.05,
    outputPATH= './data/Driver2Comm_output',
    cancer_type='Brca'
)
model.run()
```

## Downstream analysis of Driver2Comm object

### visualization of Driver-associated CCC signatures

```python
model.display_associated_FP(save=False)
```
